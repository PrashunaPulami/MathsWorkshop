1.Write a program to prompt the user for hours and rate per hour using input( ) to compute gross pay. Pay the hourly rate for the hours up to 40 and 1.5 times the hourly rate for all hours worked above 40 hours. Use 45 hours and a rate of 10.50 per hour to test the program italicised text(the pay should be 498.75).

```python

hrs = input("Enter Hours:")
h = float(hrs)
rate = input("Enter the rate per hour:")
r = float(rate)
if h <= 40:
    pay = h * r
else:
    pay = 40 * r + (h-40)*r*1.5
print("Pay", pay)
```

Q2. Write a function that compares two numbers and returns them in increasing order.

```python
def order_numbers(number1, number2):
    if number2 > number1:
        return number1, number2
    else:
        return number2, number1


smaller, bigger = order_numbers(100, 99)
print(smaller, bigger)
```

```python
#Exercise:
#A..Fibonacci Sequence Function
def fib(n):
    a = 0
    b = 1
    count = 0

    while count < n:
        print(a)
        c = a + b
        a = b
        b = c
        count += 1


n = int(input("Enter number of terms: "))
fib(n)

```

```python
#B..myPow(x, n) using while loop
def myPow(x, n):
    result = 1
    count = 0

    while count < n:
        result = result * x
        count += 1

    return result


print(myPow(2, 3))
```

```python
#C..SumDigits(n)
def SumDigits(n):
    total = 0

    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10

    return total


print(SumDigits(234))
```

```python
#D..squaredSum(list)
def squaredSum(numbers):
    total = 0

    for num in numbers:
        total = total + (num * num)

    return total


print(squaredSum([2,3,4]))
```

```python
#E..sumDigits(num)
def sumDigits(num):
    total = 0

    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10

    return total


print(sumDigits(12345))
```
